### Technical Indicators
This project utilizes three key technical indicators to analyze market trends and generate trading signals. Each indicator offers a unique perspective on price action, from trend smoothing to measuring volatility and momentum.

#### Simple Moving Average (SMA)
The Simple Moving Average (SMA) smooths out erratic price movements into a single, understandable line to reveal the underlying trend. By averaging prices over a set period, it filters out short-term market noise.

Its primary use is to generate crossover signals: a potential buy signal occurs when the price moves above the SMA, and a sell signal when it falls below.

#### Relative Strength Index (RSI)
The Relative Strength Index (RSI) is a momentum oscillator that measures the speed and change of price movements. The RSI oscillates between 0 and 100 and is primarily used to identify overbought or oversold conditions in a market.

* Overbought: An RSI reading above 70 suggests that an asset may be overvalued and is a candidate for a price pullback or reversal.

* Oversold: An RSI reading below 30 suggests that an asset may be undervalued and is a candidate for a price rally.


#### Money Flow Index (MFI)
The Money Flow Index (MFI) is a momentum indicator that uses both price and volume to measure buying and selling pressure. Often called the volume-weighted RSI, it tracks the flow of money into and out of an asset.

Because it includes volume, some traders consider its signals to be more robust. Like the RSI, it is used to identify overbought or oversold conditions.

Overbought: An MFI reading above 80 suggests buying pressure may be exhausted.

Oversold: An MFI reading below 20 suggests selling pressure may be fading.

### Optimal Strategy Analysis: Moving Average Crossover
The optimization process concluded that the most profitable strategy is a trend-following approach based on a single technical indicator: the Simple Moving Average (MA). This strategy operates on the principle of capturing gains by identifying the inception of short-term trends. It generates trading signals when the asset's price crosses over its historical average.

The specific hyperparameters identified to maximize this strategy's effectiveness are as follows:

Trading Strategy: The model exclusively uses the Simple Moving Average (MA) indicator to generate trading signals.

Number of Shares (n_shares): All trades are executed with a fixed size of 1 share. This standardizes the capital allocated to each position.

MA Window (ma_window): The moving average is calculated using a lookback period of 9 periods. This creates a responsive average that is sensitive to recent price action.

Buy Signal: A buy signal is generated when the asset's price crosses above the 9-period moving average, indicating potential upward momentum.

Sell Signal: A sell signal is generated when the asset's price crosses below the 9-period moving average, indicating a potential shift to a downtrend.

Risk Management (Long Positions):

Take Profit (tp_long_factor): A long position is automatically closed for a profit if the price increases by approximately 3.94% from the entry point (1.03936).

Stop Loss (sl_long_factor): To ensure capital preservation, a long position is automatically closed if the price drops by approximately 1.34% from the entry point (0.98661).

**Win Rate**

The Win Rate quantifies the frequency of profitable trades. As a primary indicator of reliability, it answers the question: "How consistently does the strategy generate winning trades?" While a high win rate is desirable, it can be misleading. A series of small wins can be easily erased by a single catastrophic loss, making it an incomplete metric on its own.


**Maximum Drawdown**

The Maximum Drawdown reveals the greatest percentage loss from a portfolio's peak value. It is the ultimate measure of downside risk, answering the critical question: "What is the most significant loss this strategy has historically suffered?" This metric is essential for understanding a strategy's resilience and its potential for capital preservation during adverse market conditions.
